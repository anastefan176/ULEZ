```python
# Re-read the CSV to ensure 'data' is in the expected initial state for this cell's processing
data = pd.read_csv('LaqnData (2).csv')

# Original column cleaning: strip, lowercase, replace spaces with underscores
# This will turn 'ReadingDateTime' into 'readingdatetime'
data.columns = [col.strip().lower().replace(' ', '_') for col in data.columns]
print("Columns:", data.columns.tolist())

# Rename 'readingdatetime' to 'datetime' to match the expected_cols and subsequent operations
if 'readingdatetime' in data.columns:
    data = data.rename(columns={'readingdatetime': 'datetime'})

expected_cols = ['site', 'datetime', 'value', 'units']
data = data[[col for col in expected_cols if col in data.columns]]

# Convert 'datetime' column to datetime objects
data['datetime'] = pd.to_datetime(data['datetime'], errors='coerce', dayfirst=True)
data['value'] = pd.to_numeric(data['value'], errors='coerce')

# Drop rows with NaN values in 'datetime' or 'value'
data = data.dropna(subset=['datetime', 'value'])

# Filter 'value' to be within a reasonable range
data = data[(data['value'] >= 0) & (data['value'] <= 200)]

# Extract year and month
data['year'] = data['datetime'].dt.year
data['month'] = data['datetime'].dt.month

# Filter data for years between 2020 and 2025
data = data[(data['year'] >= 2020) & (data['year'] <= 2025)]

# Create 'post_ulez' flag
data['post_ulez'] = (data['year'] >= 2023).astype(int)

# Define ULEZ map
ulez_map = {
    'BG2': 'Outer',
    'HV1': 'Outer',
    'KC5': 'Outer',
    'ST4': 'Outer',
    'IS1': 'Control',
}

# Map 'zone' based on 'site'
data['zone'] = data['site'].map(ulez_map)

# Drop rows where 'zone' is NaN (sites not in ulez_map)
data = data.dropna(subset=['zone'])

print("\nCleaned data summary:")
print(data.info())
print("\nUnique sites:", data['site'].unique())
print("\nDate range:", data['datetime'].min(), "", data['datetime'].max())
data.head()
```
